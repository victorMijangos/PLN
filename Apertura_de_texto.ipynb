Apertura y limpiado de textos

---

Aquí se muestra el código que permitirá abrir un archivo o un conjunto de archivo.

Para comenzar tomaremos un sólo archivo:

1. En el entorno de Google collab usaremos la función "*.upload" que permitirá subir archivos desde nuestra máquina a partir de un botón de selcción de archivos.
2.   El archivo subido será un objeto, tendremos que trabajar sobre el objeto.



```
from google.colab import files

corpus = files.upload()
```

Imprimiremos el archivo "celula (2).txt"

```
print corpus['celula (2).txt']
```

Ahora, trataremos de definir un lenguaje regular que contenga los signos de puntuación: punto, coma y dos puntos, y sustituirlo por el lenguaje $\epsilon$

```
import re

texto = corpus['celula (2).txt']

texto_limpio = re.sub('\.|\,|\:',r'',texto).lower()  #el comando *.lower() pasa todo el texto a minúsculas
print texto_limpio
```

Finalmente, separaremos las palabras para obtener una lista de palabras:

```
lista_palabras = texto_limpio.split()

for w in lista_palabras:
   print w
```
